#### 1.2.2 팀 경력 정리 & 합쳐주기

```python
# 소속팀이 기입된 데이터들
temp_3 = data_2[~(((data_2.team.isnull()) & (data_2.team_now.isnull())) | ((data_2.team.isnull()) & (data_2.team_now == '-')))]
temp_3
```

##### 1.2.2.1 NaN 값 정리

```python
temp_4 = temp_4.fillna('-')
temp_4
```

##### 1.2.2.2 경력 정리

```python
# 선수 팀 커리어 정리 함수
def club_career_remover(index_num):
    temp_5 = temp_4.team.iloc[index_num]
    temp_5 = temp_5.replace('임대','').replace('~','')
    temp_5 = temp_5.split('\n')
    temp_5 = list(filter(None, temp_5))

    temp_6 = []
    temp_7 = ''
    for x in range(len(temp_5)):
        if temp_5[0].startswith('1') or temp_5[0].startswith('2'):
            if x % 2 == 1:
                temp_6.append(temp_5[x])
        else:
            if x % 2 == 0:
                temp_6.append(temp_5[x])
                
        
    if temp_6 == ['-']:
        temp_6 = [temp_4.team_now.iloc[index_num]]
        
    elif temp_4.team_now.iloc[index_num] == '-':
        temp_6 = temp_6
    else:
        temp_6 = temp_6 + [temp_4.team_now.iloc[index_num]]
        
    temp_6 = list(set(temp_6))
    temp_7 = ' / '.join(s for s in temp_6)
            
    return temp_7
```

```python
# 수정
temp_4['club_career'] = '-'

for index in range(len(temp_4)):
    temp_a = club_career_remover(index)
    temp_4.club_career.iloc[index] = temp_a
    
temp_4
```

```python
temp_4.drop(columns=['team','team_now'], inplace=True)
temp_4
```

### 2. 두 데이터 합치기

```python
# 'name','overall','Stamina' 기준으로 합치기
data_merge = pd.merge(data_1,temp_4, how = 'inner', on=['name','overall','Stamina'] )
data_merge
```

```python
# 결측값 확인
print(data_merge.isna().sum().sort_values())
```

```python
# 출력 갯수 제한 해제
pd.set_option('display.max_columns',None)


# 원래대로
# pd.options.display.max_columns = 20


data_merge.head(7)
```

```python
birth = []
height = []
weight = []
star = []
sliding_tackle = []
prefer_foot = []
price_non_bp = []

for index in range(len(data_merge)):
    # 개인기 등급
    temp_num_1 = str(len(data_merge.etc_skill.iloc[index]))
    star.append(temp_num_1)
    
    # 슬라이딩 태클
    temp_num_2 = data_merge.Slide_tackle.iloc[index].split('\n')[-1]
    sliding_tackle.append(temp_num_2)
    
    # 주발 설정
    if data_merge.etc_foot.iloc[index].split(' – ')[0][-1] == data_merge.etc_foot.iloc[index].split(' – ')[1][-1]:
        prefer_foot.append('both')
    elif data_merge.etc_foot.iloc[index].split(' – ')[0][-1] > data_merge.etc_foot.iloc[index].split(' – ')[1][-1]:
        prefer_foot.append('left')
    else:
        prefer_foot.append('right')
        
    # 생일
    birth.append(data_merge.etc_birth.iloc[index].split('(')[0].rstrip())
    
    # 키
    height.append(data_merge.etc_height.iloc[index].split('c')[0])
    
    # 몸무게
    weight.append(data_merge.etc_weight.iloc[index].split('k')[0])
    
    price_non_bp.append(data_merge.price.iloc[index].split('B')[0])
```

```python
# 수정된 값으로 기입
data_merge.Slide_tackle = sliding_tackle
data_merge.etc_skill = star
data_merge.etc_foot = prefer_foot
data_merge.etc_birth = birth
data_merge.etc_height = height
data_merge.etc_weight = weight
data_merge.price = price_non_bp

data_merge.head(7)
```
